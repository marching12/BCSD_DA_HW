```python
import pandas as pd 
import numpy as np 
import matplotlib 
import matplotlib.pyplot as plt 

print("Matplotlib version", matplotlib.__version__) #현재 설치된 버전을 출력
%matplotlib inline  #그래프가 별도의 창이 아닌 노트북 내부에 표시되도록 함
%config InlineBackend.figure_format = 'retina' #그래프 출력 품질을 높임
```

```python
fig = plt.figure() #Figure 생성.
fig.suptitle('figure sample plots') #Figure의 제목 설정

fig, ax_lst = plt.subplots(2, 2, figsize=(8,5))
#Figure에 2*2=4개의 그래프를 삽입할 수 있는 공간 할당.
#각 그래프는 'ax_lst'에 배열로 저장(이차원 배열), 전체 도화지 크기를 8*5인치로 설정

ax_lst[0][0].plot([1,2,3,4], 'ro-')
#첫 번째 서브플롯, [1,2,3,4]라는 데이터를 그래프로 표현
#빨간색(r), 동그라미(o), 선(-)

ax_lst[0][1].plot(np.random.randn(4, 10), np.random.randn(4,10), 'bo--')
#두 번째 서브플롯, 정규분포를 따르는 랜덤한 데이터(크기 : x축 4*10, y축 4*10)를 배열에 저장 후 그래프로 표현.
#파란색(b), 동그라미(o), 점선(--)

ax_lst[1][0].plot(np.linspace(0.0, 5.0), np.cos(2 * np.pi * np.linspace(0.0, 5.0)))
#세 번째 서브플롯
#linsapce는 0부터 5까지의 구간을 균일한 간격으로 쪼갬
# np.cos(2 * np.pi * np.linspace(0.0, 5.0)) : 해당 x축 값에 대한 코사인 값을 계산한 y축 데이터
# '2 * np.pi' : 하나의 주기(한 바퀴)로 사용되는 각도
# 코사인 곡선의 그래프 표현

ax_lst[1][1].plot([3,5], [3,5], 'bo:')
#네 번째 서브플롯, (3,3)과 (5,5) 두 점을 이어 그래프로 표현
#파란색(b), 동그라미(o)와 점(..)

ax_lst[1][1].plot([3,7], [5,4], 'kx')
#네 번째 서브플롯, (3,5)와 (7,4) 두 점을 출력
#검은색(k) X문양(x)

plt.show() #전체 그래프 출력
```

```python
import seaborn as sns 
print("Seaborn version : ", sns._version__) #Seaborn의 버전 출력
sns.set() #Seaborn의 기본 설정을 적용
sns.set_style('whitegrid') #'whitegrid' 그래프 스타일, 흰색 배경에 격자무늬
sns.set_color_codes() #Seaborn의 색상 코드를 설정

current_palette = sns.color_palette() #현재 사용 중인 색상 팔레트 = 기본 색상 팔레트를 가져옴
sns.palplot(current_palette) #팔레트의 색상을 막대로 보여줌
```

```python
tips = sns.load_dataset("tips")
#Seaborn에서 제공하는 'tips'라는 샘플 데이터를 불러옴.
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",data=tips)
"""
relplot로 두 변수 간의 관계를 시각화
x축에 "total_bill, y축에 "tip"를 표시
smoker에 따라 데이터 포인트의 색상과 모양을 다르게 설정
사용 데이터는 "tips"
"""
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
"""
데이터 프레임 생성, 딕셔너리 자료형
time : time열에 0~500까지의 데이터 생성해 X축으로 할당
value : value열에 정규분포를 따르는 500개 난수의 누적 합을 Y축으로 할당
종류 : 선 그래프, 직접 생성한 df 사용    
"""
g.fig.autofmt_xdate()
#x축의 값을 자동으로 회전시켜 가시성 좋게 시각화
```

```python
# 펭귄의 부리의 넓이 & 길이에 따른 서식지 분포
penguin = sns.load_dataset('penguins') #penguin 샘플 데이터 로드
plt.figure(figsize=(10,5)) #10*5인치 figure 생성
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='island', style='island', s=50, data=penguin)
"""
scatterplot를 사용해 펭귄 데이터를 산포도 그래프로 시각화
x축에 'bill_length_mm', y축에 'bill_depth_mm'를 표시
'island'에 따라 데이터 포인트의 색상과 모양을 다르게 설정
s=50 : 점의 크기를 설정, 숫자가 커지면 크기가 커짐
사용 데이터는 'penguin'
"""
plt.show()
```

```python
iris = sns.load_dataset("iris")
#'iris' 샘플 데이터 로드
sns.pairplot(iris, hue='species')
"""
pairplot : 수치형 변수의 이중 관계를 나타내는 다변량 플롯을 출력
데이터를 보면 4개의 열과 하나의 종이 있으므로 4*4개 종류의 플롯을 출력
종에 따른 분포를 확인해야 할 것 같아 종 별로 색을 다르게 설정
대각선 플롯은 한 가지 변수에 대해 분포를 나타내는 그래프 출력
두 가지 변수에 대한 그래프는 산점도로 출력
"""

```

```python
g = sns.PairGrid(iris, hue='species')
#iris 데이터를 가지고 PairGrid 객체 생성, 종별로 구분짐
g.map_diag(sns.kdeplot)
#대각선에 각 변수의 분포를 커널 밀도 추정(KDE)그래프 표현
g.map_offdiag(sns.kdeplot, n_levels=6);
#비 대각선 부분을 == 두 변수 간의 관계를 KDE 등고선으로 시각화
#등고선은 밀도 값이 높을수록 세부사항이 많아짐
#n_levels=6 : 등고선의 수를 6단계로 표시.
```

```python
flights = sns.load_dataset("flights")
#'flights' 샘플 데이터 로드
flights = flights.pivot(index="month", columns="year", values="passengers")
#데이터를 표 형태로 변환
#행 month, 열 year, 셀 passengers 할당
plt.figure(figsize=(10, 10))
#10*10인치 figure 생성
ax = sns.heatmap(flights, annot=True, fmt="d")
"""
flights를 색깔로 표현하는 heatmap을 사용하여 데이터 출력
heatmap은 값이 클수록 색을 진하게 표현
annot=True : 각 셀에 값을 숫자로 표시
fmt="d" : 값을 정수로 표시
"""
```

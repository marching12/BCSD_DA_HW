```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 데이터 로드
tips = sns.load_dataset("tips")

sns.pairplot(tips, hue='smoker')
```

```python

# tips 데이터셋 로드
tips = sns.load_dataset("tips")

# 요일별 성별 손님 수 계산
guest_count = tips.groupby(['day', 'sex']).size().reset_index(name='count')

# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(data=guest_count, x='day', y='count', hue='sex', palette='viridis')

# 그래프 꾸미기
plt.title('Number of Guests by Day and Gender')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Guests')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python

# tips 데이터 불러오기
tips = sns.load_dataset('tips')

# 흡연자 데이터 필터링
smoker_data = tips[tips['smoker'] == 'Yes']

# 성별 비율 계산
gender_counts = smoker_data['sex'].value_counts()

# 데이터프레임으로 변환
gender_ratio = gender_counts.reset_index()
gender_ratio.columns = ['sex', 'count']

# 비율 계산
gender_ratio['percentage'] = (gender_ratio['count'] / gender_ratio['count'].sum()) * 100

# 그래프 그리기
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='percentage', data=gender_ratio, palette='Set2')
plt.title('Gender Ratio of Smokers')
plt.xlabel('Sex')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)
plt.show()


```

```python

# tips 데이터셋 로드
tips = sns.load_dataset("tips")

# 요일별 흡연자와 비흡연자의 팁 총합 계산 (observed 인자 추가)
total_tips_by_day_and_smoker = tips.groupby(['day', 'smoker'], observed=True)['tip'].sum().reset_index()

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(data=total_tips_by_day_and_smoker, x='day', y='tip', hue='smoker', palette='viridis')

# 그래프 꾸미기
plt.title('Total Tips by Day of the Week (Smokers vs Non-Smokers)')
plt.xlabel('Day of the Week')
plt.ylabel('Total Tips ($)')
plt.xticks(rotation=45)
plt.legend(title='Smoker', loc='upper right')
plt.tight_layout()

plt.show()
```

```python


# 데이터셋을 DataFrame으로 불러오기 (예: tips.csv)
tips = sns.load_dataset("tips")

# 팁 비율 계산
tips['tip_ratio'] = tips['tip'] / tips['total_bill'] * 100

# 요일별 팁 비율의 평균 계산
average_tip_ratio = tips.groupby('day')['tip_ratio'].mean().reset_index()

# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='tip_ratio', data=average_tip_ratio, palette='Set2')
plt.title('Average Tip Ratio by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip Ratio (%)')
plt.ylim(0, average_tip_ratio['tip_ratio'].max() + 5)  # y축 범위를 설정
plt.show()
```

```python

tips = sns.load_dataset("tips")
# 팁 비율 계산
tips['tip_ratio'] = tips['tip'] / tips['total_bill'] * 100

# pairplot 그리기
sns.pairplot(tips, hue='', vars=['total_bill', 'tip', 'tip_ratio', 'size'], palette='Set2')
plt.suptitle('Pairplot of Total Bill, Tip, Tip Ratio, and Size by Smoker Status', y=1.02)
plt.show()
```
